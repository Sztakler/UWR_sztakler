```python
import numpy
import math
import scipy.integrate as integrate
```

```python
def memoize(f):
    cache = {}

    def helper(x, y):
        if x not in cache:
            cache[x] = {}
        if y not in cache[x]:
            cache[x][y] = f(x, y)
        return cache[x][y]
    return helper
```

```python
def rombergMethod(n, a, b, f):
    def h(n):
        return (b - a) / 2**n

    def M(n):
        sumM = 0
        for i in range(1, 2**n+1):
            sumM += f(a + 0.5*(2*i - 1) * h(n))
        return h(n) * sumM

    @memoize
    def T(m, k):
        if k == 0 and m == 0:
            return (b-a) * (f(a) + f(b)) / 2
       
        if m == 0:
            return 0.5*(T(0, k-1) + M(k-1))

        return (4**m * T(m - 1, k) - T(m - 1, k - 1)) /  (4**m - 1)

    return T
```

```python
def f1(x): 
    return 2021 * x**5 - 2020 * x**4 + 2019 * x**2

def f2(x):
    return 1 / (1 + 25 * x**2)

def f3(x):
    return math.sin(7*x - 2) / x
```

```python
functions = [[f1, -1, 2], [f2, -2, 2], [f3, 2, 3*math.pi ]]
```

```python
def printRombergTable():
    n = 16
    for f, a, b in functions:
        for i in range(n+1):
            l = 3 - len(str(i))
            print(i, end= l * " " + "| ")
            for j in range(i+1):
                print(f"{rombergMethod(n, a, b, f)(j, i):4.4f}", end=" | ")
            print()
```
